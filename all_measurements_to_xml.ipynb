```python
rc_lookahead_20 = []
rc_lookahead_40 = []
rc_lookahead_60 = []

ref_1 = []
ref_5 = []
ref_9 = []

for index, row in df.iterrows():
    if row['rc_lookahead']==20:
       rc_lookahead_20.append(True) 
    else:
       rc_lookahead_20.append(False) 
    
    if row['rc_lookahead']==40:
       rc_lookahead_40.append(True) 
    else:
       rc_lookahead_40.append(False)
    
    if row['rc_lookahead']==60:
       rc_lookahead_60.append(True) 
    else:
       rc_lookahead_60.append(False)
    
    
    if row['ref']==1:
       ref_1.append(True) 
    else:
       ref_1.append(False) 
    
    if row['ref']==5:
       ref_5.append(True) 
    else:
       ref_5.append(False) 
    
    if row['ref']==9:
       ref_9.append(True) 
    else:
       ref_9.append(False) 

del df['rc_lookahead']
df.insert(9, 'rc_lookahead', True)
df.insert(10, 'rc_lookahead_20', rc_lookahead_20)
df.insert(11, 'rc_lookahead_40', rc_lookahead_40)
df.insert(12, 'rc_lookahead_60', rc_lookahead_60)

del df['ref']
df.insert(13, 'ref', True)
df.insert(14, 'ref_1', ref_1)
df.insert(15, 'ref_5', ref_5)
df.insert(16, 'ref_9', ref_9)

del df['no_asm']
#df.loc[df.no_asm == True]
df.insert(3, 'no_asm', True)

df[:5]
```

```python
additional_ft = ["size","H264", "configurationID", "usertime", "systemtime"]
```

```python
import xml.etree.cElementTree as ET

df = df.drop(columns=additional_ft)
columns = list(df.columns.values)
root = ET.Element("results")

for row in df.iterrows():
    config=""
    
    for i in columns:
        if (df[i][row[0]]==1):
            config=config+i+","
    
    doc = ET.SubElement(root, "row")

    ET.SubElement(doc, "data", columname="Configuration").text = config
    ET.SubElement(doc, "data", columname="Performance").text = str(df["elapsedtime"][row[0]])

tree = ET.ElementTree(root)
tree.write("measurements.xml")
```
